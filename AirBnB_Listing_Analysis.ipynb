# **AirBnB Listing Analysis**

## **Objective 1: Data Profile & QA**


```python
#importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the data from csv file
listings = pd.read_csv('Listings.csv', encoding='ISO-8859-1', low_memory=False)
```

```python
listings.head()
```

```python
listings.shape
```

```python
listings.info()
```

```python
#changing the data format to datetime for a specific column
listings['host_since'] = pd.to_datetime(listings['host_since'])
```

```python
# choosing specific columns for analysis from the listings dataframe
paris_listings = listings.query("city == 'Paris'").loc[:, ["host_since", "neighbourhood", "city", "accommodates", "price"]]
```

```python
paris_listings.info()
```

```python
#checking for null values
paris_listings.isnull().sum()
```

```python
#removing the null values
paris_listings = paris_listings.dropna()
```

```python
paris_listings.isnull().sum()
```

```python
paris_listings.describe()
```

```python
# the minimum value of accomodates is 0 so checking for those records
paris_listings.query("accommodates == 0").count()
```

```python
## the minimum value of price is 0 so checking for those records
paris_listings.query("price == 0").count()
```

```python
paris_listings.query("accommodates == 0 and price == 0").count()
```

## **Objective 2: Prepare the data for visualization**


```python
paris_listings_neighbourhood = paris_listings.groupby("neighbourhood").agg({"price" : "mean"}).sort_values("price")
```

```python
paris_listings_neighbourhood.head()
```

```python
paris_listings_accomodations = paris_listings.query("neighbourhood == 'Elysee'").groupby("accommodates").agg({"price": "mean"}).sort_values("price")
```

```python
paris_listings_accomodations.head()
```

```python
paris_listings_over_time = paris_listings.set_index("host_since").resample("Y").agg({"neighbourhood" : "count", "price" : "mean"})
```

```python
paris_listings_over_time.head()
```

## **Objective 3 : Visualize the data and summarize findings**


```python
paris_listings_neighbourhood.plot.barh()
```

```python
paris_listings_neighbourhood.plot.barh()

sns.despine()
```

```python
paris_listings_neighbourhood.plot.barh(title="Average Listing Price by Neighbourhood in Paris", xlabel="Price Per Night (Euros)", ylabel="Neighbourhood", legend=None)

sns.despine()
```

```python
paris_listings_accomodations.plot.barh(title="Average Listing Price by Accomodation Number", xlabel="Price Per Night (Euros)", ylabel="Accomodation Capacity", legend=None)

sns.despine()
```

```python
paris_listings_over_time["neighbourhood"].plot(ylabel="New Hosts", title="New AirBnB Hosts in Paris Over Time")
sns.despine()
```

```python
paris_listings_over_time["price"].plot(ylabel="Average Price (Euros)", title="Average AirBnB Price in Paris Over Time")
sns.despine()
```

```python
fig, ax = plt.subplots()
ax.plot(paris_listings_over_time.index, paris_listings_over_time["neighbourhood"], label="New Hosts",c="pink")

ax.set_ylabel("New Hosts")

ax2 = ax.twinx()
ax2.plot(paris_listings_over_time.index, paris_listings_over_time["price"], label="Average Price")
ax2.set_ylim(0)
ax2.set_ylabel("Average Price")

ax.set_title("2015 Regulations Lead to Fewer New Hosts and Higher Prices")
```
